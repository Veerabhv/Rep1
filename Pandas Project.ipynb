Need to convert Award Amount Column to INtegers or Floats. the only non-numerical charecter in Amounts is $. By removing the $ sign it should allow defining floats. 

```python
philafunds['2010/11/Award Amount']= philafunds['2010/11/Award Amount'].str.replace('$','')
philafunds.head()
```

```python
philafunds['2010/11/Award Amount']= philafunds['2010/11/Award Amount'].astype(float)
philafunds.head()
```
