#### Since the analysis is based on funding, we need to convert Award Amount Column to Integers or Floats. The only non-numerical charecter in Amounts column is "Dollar" sign. By removing the "Dollar " sign we should be able to define floats. 

```python
fundingdata['2010/11/Award Amount']= fundingdata['2010/11/Award Amount'].str.replace('$','')
```

```python
fundingdata['2010/11/Award Amount']= fundingdata['2010/11/Award Amount'].astype(float)
fundingdata.head()
```

```python
fundingdata.describe()
```

```python
fundingdata.dtypes
```

#### The items in Amounts column are now float objects. The describe()function above shows the statistics - including mean and Std Deviation. However the summary shows only 757 counts for funding amounts and not 831 that are needed. It is likely that some of the funding amount cells are nan values.

#### Delete nan values in the '2010/11/Award Amount' column. dropna command can be used to remove nan values in the columns specified. We can do this removal in place to keep the operation simpler.

##### Syntax: df.dropna(subset = ['column1_name', 'column2_name', 'column3_name'], inplace=True)

```python
fundingdata.dropna(subset=['2010/11/Award Amount'],inplace=True)
```

#### Check the data again to see if the nan values are removed.

```python
fundingdata.describe()
```

#### The nan values have been removed and we are left with 757 records (instead of 831), across 46 states (instead of 50).

### To conduct further analysis, we need to understand the background infomration about the funding data.
#### According to the Metadata provided for "School Improvement 2010 Grants": States were awarded nearly $3.5 billion in School #### Improvement Grant funds in 2010 to turn around their persistently lowest achieving schools. One of the following four models were to be implemented: Turnaround Model, Restart Model, School Closure and Transformation Model.
#### Only one column has numerical values and all others are strings, plotting a scatter graph may not yield much insights.
#### We can start with a bar graph for - funding by district 

```python
trace1 = [go.Bar(x=fundingdata.District Name, y=fundingdata.2010/11/Award Amount)]
layout = go.Layout(title="Award Amounts by School Districts", yaxis=dict(title='Amounts'), xaxis=dict(title='School Districts'))
fig = go.Figure(data=trace1, layout=layout)
py.offline.iplot(fig)
```
